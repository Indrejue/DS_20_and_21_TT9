```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.neighbors import NearestNeighbors
pd.set_option('display.max_columns', None)
```

```python
weed = pd.read_csv('https://raw.githubusercontent.com/Indrejue/DS_20_and_21_TT9/main/Data/weed.csv', index_col='name')
```

```python
weed.head()
```

```python
weed = weed.drop_duplicates()
```

```python
weed.reset_index(level=0, inplace=True)

weed.head()
```

```python
y_set = ['name', 'brand', 'strain', 'rating', 'description', 'type_indica', 'type_hybrid', 'type_sativa']
weed_data = weed.drop(y_set, axis=1)
weed_target = weed[y_set]
```

```python
weed_data.tail()
```

```python
weed_data.shape
```

```python
weed_target.tail()
```

```python
nn = NearestNeighbors(n_neighbors = 5, algorithm = 'kd_tree')

nn.fit(weed_data)
```

```python
out = nn.kneighbors([weed_data.iloc[35].values])
```

```python
indexs = out[1].flat[0:5].tolist()
```
