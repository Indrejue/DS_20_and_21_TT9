```
# this is the dataset cleaned by Jacob and what the team went forward with!
# descriptions, ailments, flavors and effects have been parsed into columns
# ratings have also been incorporated
weed_jacob = pd.read_csv(weed_jacob_link)
print(weed_jacob.shape)
weed_jacob.head()
```

```
pd.set_option('display.max_rows', 120)
```

```
# 20 effects
# 3 types
# 8 ailments
# for simple analysis purposes, we'll stick with ailments and the top 3 
# categories
weed_jacob.astype(bool).sum(axis=0)
```

```
# let's create a new dataset to play around with and leave the original untouched
weed_jacob_type = weed_jacob[['category_wax','category_vapes','category_oil','type_indica', 'type_hybrid', 'type_sativa','rating']].copy()
```

```
# how many each? what does the distribution look like?
weed_jacob_type.astype(bool).sum(axis=0)
```

```
weed_jacob_type.shape
```

```
# if we want to play around with these we need to convert these to strings (obj)
weed_jacob_type.dtypes
```

```
weed_jacob_type['type_indica'] = weed_jacob_type['type_indica'].astype(str)
weed_jacob_type['type_hybrid'] = weed_jacob_type['type_hybrid'].astype(str)
weed_jacob_type['type_sativa'] = weed_jacob_type['type_sativa'].astype(str)
weed_jacob_type['category_wax'] = weed_jacob_type['category_wax'].astype(str)
weed_jacob_type['category_vapes'] = weed_jacob_type['category_vapes'].astype(str)
weed_jacob_type['category_oil'] = weed_jacob_type['category_oil'].astype(str)
```

```
weed_jacob_type.dtypes
```

```
# this is how we start to merge for exploratory data analysis
weed_jacob_type['type_indica'] = weed_jacob_type['type_indica'].replace('1','Indica')
weed_jacob_type['type_hybrid'] = weed_jacob_type['type_hybrid'].replace('1','Hybrid')
weed_jacob_type['type_sativa'] = weed_jacob_type['type_sativa'].replace('1','Sativa')
weed_jacob_type['category_wax'] = weed_jacob_type['category_wax'].replace('1','Wax')
weed_jacob_type['category_vapes'] = weed_jacob_type['category_vapes'].replace('1','Vapes')
weed_jacob_type['category_oil'] = weed_jacob_type['category_oil'].replace('1','Oil')
```

```
# this makes joining clean and consistent
weed_jacob_type['type_indica'] = weed_jacob_type['type_indica'].replace('0','')
weed_jacob_type['type_hybrid'] = weed_jacob_type['type_hybrid'].replace('0','')
weed_jacob_type['type_sativa'] = weed_jacob_type['type_sativa'].replace('0','')
weed_jacob_type['category_wax'] = weed_jacob_type['category_wax'].replace('0','')
weed_jacob_type['category_vapes'] = weed_jacob_type['category_vapes'].replace('0','')
weed_jacob_type['category_oil'] = weed_jacob_type['category_oil'].replace('0','')
```

```
weed_jacob_type.tail(50)
```

```
# we only want to analyze observations with a rating, since that's something we 
# can tangibly work with
weed_jacob_type_ratings = weed_jacob_type.dropna()
```

```
weed_jacob_type_ratings.shape
```

```
# double check that ratings are the only rows in this dataset
weed_jacob_type_ratings.head(30)
```

```
#turn off SettingWIthCopyWarning to allow merging to occur
pd.set_option('mode.chained_assignment', None)
weed_jacob_type_ratings['Top Category'] = weed_jacob_type_ratings[['category_wax','category_vapes','category_oil']].apply(lambda x: ' '.join(x), axis = 1)
```

```
weed_jacob_type_ratings['Top Type'] = weed_jacob_type_ratings[['type_indica','type_hybrid','type_sativa']].apply(lambda x: ' '.join(x), axis = 1)
```

```
weed_jacob_type_ratings.tail(30)
```

```
# the final merged file that's easy to manipulate and relatively simplified for
# the most popular classifications within each column
final = weed_jacob_type_ratings[['rating','Top Category','Top Type']].copy()
final.head()
```

```
# although these aren't normalized between classifications, the absolute number 
# is significant in demonstrating which types of products are the most popular
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

sns.countplot(hue = final["Top Type"], palette = "Set1", x = final["rating"])
plt.title("Distribution of Product rating According to Strain Type")
plt.show()
```

```
figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

sns.countplot(hue = final["Top Category"], palette = "Set1", x = final["rating"])
plt.title("Distribution of Product rating According to Consumption Category Type")
plt.show()
```

```
x,y = 'rating', 'Top Category'

(final
.groupby(x)[y]
.value_counts(normalize=True)
.mul(100)
.rename('percent')
.reset_index()
.pipe((sns.catplot,'data'), x=x,y='percent',hue=y,kind='bar'))

figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
```
