```python
prod_data = pd.read_csv('https://raw.githubusercontent.com/kushyapp/cannabis-dataset/master/Dataset/Products/products-kushy_api.2017-11-14.csv')
strain_data = pd.read_csv('https://raw.githubusercontent.com/kushyapp/cannabis-dataset/master/Dataset/Strains/strains-kushy_api.2017-11-14.csv')
rate = pd.read_csv('https://raw.githubusercontent.com/Indrejue/DS_20_and_21_TT9/main/Data/cannabis.csv')
```

```python
def cleanup(df):
    df.drop(columns=['id', 'slug'], inplace=True)
    # drops columns with a lot of nan values greater than half
    bad_cols = [col for col in df.columns
                if df[col].count() < 500]
    df.drop(columns=bad_cols, inplace=True)    
    return df
```

```python
prod_data = cleanup(prod_data)
strain_data = cleanup(strain_data)
```

```python
def strain_clean(df):
    # remove special chars
    df['strain'] = df['strain'].str.replace('-', ' ')
    # lowercase the words
    df['strain'] = df['strain'].str.lower()
    # remove leading and trailng whitespaces
    df['strain'] = df['strain'].str.strip()
    return df
```

```python
strain_data.rename(columns = {'name':'strain'}, inplace = True)
rate.rename(columns = {'Strain':'strain'}, inplace = True)
prod_data = strain_clean(prod_data)
strain_data = strain_clean(strain_data)
rate = strain_clean(rate)
plant = pd.merge(prod_data, strain_data, how='left', on=['strain'])
plant.tail(30)
```
